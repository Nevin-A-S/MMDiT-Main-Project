```python
import os

# Set the folder path
folder_path = "/media/sct/projects/MMDit/Project/MMDiT-Main-Project/dataset/MRI_DATASET/Large_Tumor_dataset_processed"

# Iterate through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".png"):  # Check for PNG files
        file_path = os.path.join(folder_path, filename)
        if os.path.getsize(file_path) < 1024*10:  # Check if file size is less than 1 kB
            os.remove(file_path)  # Delete the file
            print(f"Deleted: {filename}")

print("Processing complete.")
```

```python
import os
import pandas as pd

# Set the folder path and CSV file path
folder_path = "/media/sct/projects/MMDit/Project/MMDiT-Main-Project/dataset/MRI_DATASET/Large_Tumor_dataset_processed"
csv_file_path = "/media/sct/projects/MMDit/Project/MMDiT-Main-Project/dataset/MRI_DATASET/caption_processed_tumor.csv"

# Load the CSV file
df = pd.read_csv(csv_file_path)

# Keep only entries where the file still exists in the folder
df = df[df['filename'].apply(lambda x: os.path.exists(os.path.join(folder_path, x)))]

# Save the updated CSV
df.to_csv("/media/sct/projects/MMDit/Project/MMDiT-Main-Project/dataset/MRI_DATASET/caption_large_cleaned_tumor.csv", index=False)

print("CSV file updated. Deleted file entries removed.")
```
