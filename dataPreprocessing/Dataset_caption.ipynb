```python
import pandas as pd
```

```python
alzheimers = pd.read_csv("Dataset caption-20250223T062943Z-001/Dataset caption/Final CSVs/alzheimers_hf_dataset_info.csv")
```

```python
def replace_alzheimers(x):
    if x == '0':
        return 'Brain MRI of a patient with Mild Demented alzheimers'
    elif x == '1':
        return 'Brain MRI of a patient with Moderate Demented alzheimers'
    elif x == '2':
        return 'Brain MRI of a patient with Non Demented alzheimers'
    elif x == '3':
        return 'Brain MRI of a patient with Very Mild Demented alzheimers'
    else:
        return "Brain MRI of a patient without any disease"
```

```python
alzheimers['label'] = alzheimers['label'].apply(replace_alzheimers)
```

```python
alzheimers.drop(labels=["filename"],axis=1, inplace=True)
```

```python
brain_tumour_hf = pd.read_csv("Dataset caption-20250223T062943Z-001/Dataset caption/Final CSVs/brain_tumour_hf_dataset_info.csv")
```

```python
def brain_tumour_hf_labels(x):
    if x == 0:
        return 'Brain MRI of a patient with Giloma tumour'
    elif x == 1:
        return 'Brain MRI of a patient with Meningioma tumour'
    elif x == 2:
        return 'Brain MRI of a patient without any disease'
    elif x == 3:
        return 'Brain MRI of a patient with Pituitary tumour'
    else:
        return "Brain MRI of a patient without any disease"
```

```python
brain_tumour_hf['label'] = brain_tumour_hf['label'].apply(brain_tumour_hf_labels)
```
